```python
df = pd.read_csv('/home/hasitha/PycharmProjects/DSS-Framework/output/rain.csv',header=0)
df['time'] = pd.to_datetime(df['time'])
df
```

```python
resample_df = (df.set_index('time').resample('5T').first().reset_index().reindex(columns=df.columns))
resample_df
```

```python
interpolate_df = resample_df.interpolate(method ='linear', limit_direction ='forward').set_index(keys='time')
interpolate_df
```
