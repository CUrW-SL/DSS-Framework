```python
template = open(basin_template_file, 'r') 
lines = template.readlines()
line_count = 1
with open(basin_file, 'w') as actual:
    for line in lines: 
        if line_count == 45:
            new_line = '     Initial Baseflow: {}\n'.format(50)
        elif line_count == 95:
            new_line = '     Initial Baseflow: {}\n'.format(30)
        elif line_count == 128:
            new_line = '     Initial Baseflow: {}\n'.format(25)
        elif line_count == 187:
            new_line = '     Initial Baseflow: {}\n'.format(50)
        elif line_count == 219:
            new_line = '     Initial Baseflow: {}\n'.format(5)
        else:
            new_line = line
        actual.write(new_line)
        line_count+=1
```

```python
import geopandas as gpd
sub_catchment_shape_file = '/home/hasitha/PycharmProjects/distributed_hechms/resources/sub_catchments/sub_subcatchments.shp'
```

```python
catchment_df = gpd.GeoDataFrame.from_file(sub_catchment_shape_file)
catchment_df
```

```python
total_area = 0
for index, row in catchment_df.iterrows():
    total_area += row['Area']
print('total_area : ', total_area)
```

```python
area_ratio = {}
for index, row in catchment_df.iterrows():
    area_ratio[row['Name_of_Su']] = row['Area']/total_area
print('area_ratio : ', area_ratio)
```
