```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
input_path = '/home/hasitha/PycharmProjects/DSS-Framework/input'
fcst_file_path = os.path.join(input_path,'Janakala_Kendraya_fcst.csv')
obs_file_path = os.path.join(input_path,'Janakala_Kendraya_obs.csv')
```

```python
fcst_df = pd.read_csv(fcst_file_path)
fcst_df = fcst_df.rename(columns={"value": "fcst"})
fcst_df.plot.line(x='time', y='fcst')
```

```python
fcst_df.set_index('time', inplace=True)
print(fcst_df)
```

```python
obs_df = pd.read_csv(obs_file_path)

obs_df = obs_df.rename(columns={"value": "obs"})
obs_df.plot.line(x='time', y='obs')
```

```python
#obs_df.set_index('time', inplace=True)
obs_df
```

```python
merge_df = pd.merge(obs_df, fcst_df, on='time')
#merge_df.set_index('time', inplace=True)
merge_df
```

```python
merged_df.plot.line(x='time', y=['obs','fcst'],figsize=(50,20))
```

```python
ax = merged_df.plot(x="time",kind="line")
merged_df.plot(x="time", y="obs", kind="line", ax=ax, color="C2")
merged_df.plot(x="time", y="fcst", kind="line", ax=ax, color="C3")
plt.show()
```
